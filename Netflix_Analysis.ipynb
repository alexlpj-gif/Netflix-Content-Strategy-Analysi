```python
import pandas as pd
import matplotlib.pyplot as plt

# 讀取數據
df = pd.read_csv('Netflix.csv')

# 觀察資料規模與欄位
print(f"資料筆數: {df.shape[0]}")
print(df.info())
```

```python
# 1. 計算 Movie 與 TV Show 的數量
type_counts = df['type'].value_counts()

# 2. 設定畫布比例與字體
plt.figure(figsize=(7, 7))

# 3. 畫出圓餅圖 (使用 Netflix 標誌性的紅與黑)
# #E50914 是 Netflix 紅, #221F1F 是深灰色
colors = ['#E50318', '#000000']

plt.pie(type_counts, 
        labels = type_counts.index, 
        autopct = '%1.1f%%', 
        startangle = 140, 
        colors = colors,
        explode = (0.05, 0), # 讓 Movie 稍微突出一點點，增加視覺層次
        textprops={'color':"white", 'weight':'bold', 'fontsize':12})

# 4. 加上商業標題
plt.title('Netflix Content Strategy: Movie vs TV Show', fontsize=15, pad=20)

# 5. 顯示圖表
plt.show()
```
